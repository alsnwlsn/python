```python
t = np.arange(0,12,1)
y1=np.sin(t)
y2=np.cos(t)
```

```python
plt.plot(t,y1,'--x',color='lightblue')
plt.plot(t,y2,'-o',marker='+',color='blue')
plt.grid()
plt.figure(figsize=(12,6))
```

```python
plt.subplot(1,2,1)
plt.plot(t,y1,'--x',color='lightblue')
plt.subplot(1,2,2)
plt.plot(t,y2,'--+',color='blue')
```

subplot 을 이용해서 2행 1열 구조 2개씩
4가지 색상 다르게
4가지 스타일 다르게
4가지 마커 서로 다르게
범례 표시
x,y,제목

```python
x = np.arange(0,5,1)
y1=x+1
y2=x+2
y3=x+3
y4=x+4
plt.subplot(1,2,1)
plt.plot(x,y1,'--x',color='blue')
plt.plot(x,y2,'-*',color='lightblue')
plt.xlabel("vision")
plt.ylabel("human")
plt.title("human vision")
plt.legend('01')
plt.subplot(1,2,2)
plt.plot(x,y3,'-h',color='red')
plt.plot(x,y4,'-d',color='lightgreen')
plt.xlabel("vision")
plt.ylabel("human")
plt.title("human vision")
plt.legend('23')
```

```python
import pandas as pd
p = 'C.csv'
p
```

```python
df=pd.read_csv(p,encoding="cp949")
df
```

```python
a=df['이름']
b=df['우뽈락']
c=df['김멍게']
d=df['신해삼']
e=df['임방어']
f=df['전연어']
g=df['홍아귀']
```

```python
from matplotlib import rc
rc('font', family='gulim')
plt.plot(a,b,'--x',color='dodgerblue')
plt.plot(a,c,'--h',color='blue')
plt.plot(a,d,'--d',color='red')
plt.plot(a,e,'--p',color='k')
plt.plot(a,f,'-->',color='lightgreen')
plt.plot(a,g,'--<',color='orange')
plt.xlabel('과목')
plt.ylabel('성적')
plt.title('학생성적')
plt.legend('우김신임전홍')
```
