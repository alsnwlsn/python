```python
import matplotlib.pyplot as plt
a=[10, -20, 30, 40]
b=['kor','eng','math','info']
c=[100, -50, 60, 80]
plt.bar(b,a,color="lightblue",)
plt.barh(b,c,color="darkblue",)
plt.ylabel('jumsu')
plt.xlabel('subject')
plt.title('score')
plt.legend('value')
```
