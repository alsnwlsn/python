```
import matplotlib.pyplot as plt
import csv
f = open('/content/drive/My Drive/Data/ta.csv')
data = csv.reader(f)
next(data)
mx_temp = []
for row in data : 
  if row[-1] !='':
    mx_temp.append(float(row[-1]))
plt.plot(mx_temp)
plt.show()
```
