```
import matplotlib.pyplot as plt
import csv
f = open('/content/drive/My Drive/Data/Incheon_temp.csv')
data = csv.reader(f)
next(data)
mx_temp = []
for row in data :
  if row [-1] != '' and row[0].split('-')[1] =='11' and row[0].split('-')[2] =='11':
    mx_temp.append(float(row[-1]))
plt.plot(mx_temp)
plt.show
```

```
f = open('/content/drive/My Drive/Data/201811_201811__.csv')
data = csv.reader(f)
mytown = []
for row in data :
  if '신도림' in row[0] :
    for i in range(3, 104) :
      mytown.append(int(row[i]))
plt.figure(dpi=300)
plt.style.use('ggplot')
plt.title('sindorim')
plt.plot(mytown)
plt.show
```
