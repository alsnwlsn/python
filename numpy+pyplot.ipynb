```python
t = np.arange(0,12,1)
y1=np.sin(t)
y2=np.cos(t)
```

```python
plt.plot(t,y1,'--x',color='lightblue')
plt.plot(t,y2,'-o',marker='+',color='blue')
plt.grid()
plt.figure(figsize=(12,6))
```

```python
plt.subplot(1,2,1)
plt.plot(t,y1,'--x',color='lightblue')
plt.subplot(1,2,2)
plt.plot(t,y2,'--+',color='blue')
```

```python
x = np.arange(0,5,1)
y1=x+1
y2=x+2
y3=x+3
y4=x+4
plt.subplot(1,2,1)
plt.plot(x,y1,'--x',color='blue')
plt.plot(x,y2,'-*',color='lightblue')
plt.xlabel("vision")
plt.ylabel("human")
plt.title("human vision")
plt.legend('01')
plt.subplot(1,2,2)
plt.plot(x,y3,'-h',color='red')
plt.plot(x,y4,'-d',color='lightgreen')
plt.xlabel("vision")
plt.ylabel("human")
plt.title("human vision")
plt.legend('23')
```
