```
import numpy as np
from PIL import Image
from wordcloud import WordCloud
```

```
img = '/content/drive/My Drive/Data/image.jpg'
text = open('/content/drive/My Drive/Data/painkiler.txt').read()
text
```

```
Image.open(img)
```

```
Ip_mask = np.array(Image.open(img))
Ip_mask
```

```
Ip_wc=WordCloud(background_color = 'white',mask=Ip_mask,max_font_size=50,contour_width=5,contour_color='green').generate(text)
```

```
Ip_wc.words_
```

```
Ip_wc.to_image()
```
