```
prince=[]
for line in file:
  line=line.strip()
  line=line.lower()
  word=line.split()
  prince.extend(word)
file.close
```

```
print(prince)
```

```
def count(d):
  freq = {}
  for i in d:
    if i not in freq:
      freq[i]=1
    elif i in freq:
      freq[i]+= 1
  return freq
```

```
lp=count(prince)
lp
```

```
freq = sorted(lp.items(), key = lambda i :i[1], reverse=True)
freq[:30]
```

```
freq=dict(freq)
freq
```

```
new_dic = {}
for key in freq:
  if len(key)>3:
    new_dic[key]=freq[key]
print(new_dic)
```

```
word=list(new_dic.keys())
word
```

```
cnt=list(new_dic.values())
max(cnt)
```

```
import matplotlib.pyplot as plt
plt.grid()
plt.plot(cnt)
```
